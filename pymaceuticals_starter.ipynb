## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress
from scipy.stats import sem

# Study data files
mouse_metadata_path = "Mouse_metadata.csv"
study_results_path = "Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
mouse_results = pd.merge(mouse_metadata, study_results, how='outer', on='Mouse ID')


# Display the data table for preview
mouse_results
```

```python
# Checking the number of mice.
mice_count = mouse_results['Mouse ID'].nunique()
mice_count
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 

duplicate_mouse = mouse_results[mouse_results.duplicated(['Mouse ID', 'Timepoint'])] 
duplicate_mouse
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mouse = mouse_results.loc[mouse_results['Mouse ID'] == 'g989']
duplicate_mouse
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_results = mouse_results.drop_duplicates(subset=['Mouse ID', 'Timepoint'], keep=False)
clean_results
```

```python
# Checking the number of mice in the clean DataFrame.
clean_mouse_count = clean_results['Mouse ID'].nunique()
clean_mouse_count
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
x_values = clean_results[clean_results['Drug Regimen'] == 'Capomulin']['Weight (g)']   
y_values = clean_results[clean_results['Drug Regimen'] == 'Capomulin']['Tumor Volume (mm3)']  
plt.scatter(x_values,y_values)
plt.xlabel('Weight (g)')
plt.ylabel('Tumor Volume (mm3)')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
correlation = round(st.pearsonr(x_values, y_values)[0],2)
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(0,50),fontsize=15,color="red")
plt.xlabel('Weight (g)')
plt.ylabel('Tumor Volume (mm3)')
print(f"The r-squared is: {rvalue**2}")
print(f"The correlation coefficient between weight and tumor volume is {correlation}")
plt.show()
```
